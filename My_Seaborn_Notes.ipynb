*Distribution plot* --
To discuss different plot type with seaborn that allows us to visualize the distibution of Data set

```python
tips = sns.load_dataset('tips') # Is an built in dataset
tips.head()
```

## DISTRIBUTION PLOT

It allows us to show univariate set of observation - Univariate meaning- one vaiable quantity.
we pass single column of the dataset

```python
#sns.distplot(tips['total_bill']) # we get a HISTOGRAM and KDE - Kernel Density Estimatior

#sns.distplot(tips['total_bill'],kde=False) #there is no Kernel Dessity Estimator

sns.distplot(tips['total_bill'],kde=False,bins = 30) # setting up bins


# we can over laps the plots
```

## JOINT PLOTS

allows us to match up with DIST PLOT with bivariate data, can combine two different distibution plot 

```python
sns.jointplot(x='total_bill',y='tip',data=tips) # X and Y are the column name - the two things which you want to compare
#scatter plot is by default in the in the joint plot
```

1. The "Kind" argument allows us to make a hexagon
2. It is darker where we have more data points and ligher where we have less data points
3. Basically showing distribution based on hexagon

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')


```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
#the "reg" keyword in puts a regression line on the scatter plot
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

kind='kde' shows two dimentions KDE , it shows density where these points match up the most.

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='resid')
```

kind when set to resid will plot the residuals of a regression, essesntially when Y is regressed on X

## Categorical plots

We will be focused on plotting column such as
1. This sex, gender, or smoker, daytime in reference either to one of the numerical columns.
2. X is the categorial column and Y is the numerical column
3. Data is the dataframe which we can use

#Violin plot
1.Play a very similar role to a boxplot
2.A violin plot is a method of plotting numeric data. It is similar to a box plot, with the addition of a rotated kernel density plot on each side. Violin plots are similar to box plots, except that they also show the probability density of the data at different values, usually smoothed by a kernel density estimator.

```python
sns.violinplot(x ='day',y ='total_bill',data=tips)
#sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y ='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y ='total_bill',data=tips,hue='smoker',split = True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
# what is a stip plot?
# Stip plot is like a scatter plot with only one extention.
#It is best used when the data set in not too large.
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter =True) # added more density 
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter =True,hue ='smoker')
#sns.stripplot(x='day',y='total_bill',data=tips,jitter =True,hue ='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter =True,hue ='sex',split=True)
```

Combine strip plot with violin plot we get swarm plot


Draw a categorical scatterplot with non-overlapping points. This function is similar to stripplot() , but the points are adjusted (only along the categorical axis) so that they don't overlap. #looks good but not good for large number of data 

```python
sns.swarmplot(x='day',y='total_bill',data=tips)

```

1.we can combine them together to do a good data analysis.

2.These plots are good at for doing expplanatory analysis but not good for explaining it to executive.

3.Most Executives would like to see more familiar plots like scatter,box plots etc

```python
sns.violinplot(x='day',y ='total_bill',data=tips,hue='smoker',split = True)
sns.swarmplot(x='day',y='total_bill',data=tips,color ='black')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')

#Factor plot is informative when we have multiple groups to compare
```

# Matrix plot

```python
flights = sns.load_dataset('flights')
flights.head() #number of passengers flights in a given month and year
```

1.Heat plots is our primary way of showing some sort of matrix plot

2.For heat plot to work propertly the data should be already in matrix 

3.That is the cell value; the index name and the column name should match up so that the cell can indicate relavannce to both of the names

```python

tc =tips.corr()
sns.heatmap(tc)
```

```python
sns.heatmap(tc,annot=True) #annotates values in the cell
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm') #annotates values in the cell
```

```python
flights
fc=flights.pivot_table(index='month',columns ='year',values='passengers') # making the flight data to a heat map
```

```python
sns.heatmap(fc)
```

```python
sns.heatmap(fc,cmap='coolwarm',linecolor='white',linewidths=1) #makes it more easier to read
```

# Cluster map

Cluster maps are putting the similar groups close to each other

```python
sns.clustermap(fc)
```

```python
sns.clustermap(fc,cmap='coolwarm')
```

```python
sns.clustermap(fc,cmap='coolwarm',standard_scale=1) #normalized versiop of heat map
```

# GRID PLOT

This is used to automate subplot based of features.

we will work on iris data set

```python
iris =sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris,hue='species')
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
g =sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```
