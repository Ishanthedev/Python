```python
'''
Q1: The attached carprices.txt contains a single column of car prices in 1000’s. 
    Read this data into your program.  The first line is the heading and should be ignored. 
    For any line containing non-numeric data, raise a ValueError exception and print a message saying,
    ‘This data is invalid and will be ignored’ together with the exception object.
    You should divide each number by 10 and round it down to the nearest integer. 
    Then plot a histogram that looks like the one shown below.  
    The number of bins should be chosen so as to produce the graph shown below.
'''
import matplotlib.pyplot as plt
import math as m
list_1 =[]
list_2 =[]
list_3 =[]

carprice = open("carprices.txt","r")
carpice_ln =carprice.readlines()

for x in carpice_ln:
    y = x.strip('\n')
    if y not in list_2:
        if y == 'Type':
            continue
        list_2.append(y)
for x in list_2:
    try:
        list_3.append(m.floor(int(x)/10))
    except ValueError as e:
        print("Data is invalid and will be ignored!", e)

bins = 50
plt.hist(list_3,bins,color='blue',histtype = 'bar',rwidth = 0.6)
plt.xlabel('price in ten of thousand')
plt.ylabel('Number of cars')
plt.title('Frequency of Car Prices')
plt.show()

carprice.close()
```

```python
'''
Q2: The attached cartype.txt contains a single columns listing the type of car (van, compact, etc). 
Read this data into your program.  The first line is the heading and should be ignored.  
Then plot a piechart after computing the frequency for each car type. 
Your chart should look like the one shown below. Each slice should be separated from the next as shown below.
'''
import matplotlib.pyplot as plt
list_1 =[]
list_2 =[]
list_3 =[]

small = 0
midsize= 0
compact = 0
large = 0
sporty = 0
van = 0

cartype = open('cartype.txt')
ct_lines = cartype.readlines()  

for x in ct_lines:
    y = x.strip('\n')
    if y  not in list_2:
        if y == 'Type':
            continue
        list_2.append(y)
        
for x in ct_lines:
    y = x.strip('\n')
    list_3.append(y)
    small = list_3.count('small')
    midsize = list_3.count('midsize')
    compact = list_3.count('compact')
    large = list_3.count('large')
    sporty = list_3.count('sporty')
    van = list_3.count('van')

Frequency = [small,midsize,compact,large,sporty,van]
colors = ['m','b','m','c','c','b']

plt.pie(Frequency, labels = list_2, colors = colors, explode =(0.025,0.025,0.025,0.025,0.025,0.025),startangle = 90,
       radius = 1.2, autopct ='%2.2f%%')

plt.show()    
    
cartype.close()
```

```python
'''
Q3:	The attached wildlife.txt contains data about the population of whales, dolphins and whales over a number of years.
Read this data into your program.  The first line is the heading and should be ignored.
Then plot a line charts displaying the population trend for each animal. 
Your graph should look like the one shown below.  

'''

import matplotlib.pyplot as plt
years =[]
bear =[]
dolphin =[]
whale =[]

wildlife = open('Wildlife.txt')
wline= wildlife.readlines()
for x in wline:
    y = x.split('\t')
    
    try:
        years.append(int(y[0]))
        bear.append(int(y[1]))
        dolphin.append(int(y[3]))
        whale.append(int(y[5]))
    except Exception:
        pass
plt.plot(years,bear,label ='Bears') 
plt.plot(years,dolphin,label ='dolphin')
plt.plot(years,whale,label ='whale')
plt.xlabel('Years')
plt.ylabel('Wildlfe Poplulation')
plt.title('Wildlife Population Per Year')
plt.legend(loc = 'upper left')

plt.show()
plt.close()
wildlife.close()
```
